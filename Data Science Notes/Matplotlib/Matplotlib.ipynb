# 1. Bar Plot or Linear Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = ["Python", "C", "C++", "Java"]
y = [85,73,60,82]
plt.xlabel("Languages", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("MoonSolution",size=20)
# c = ["g", "b", "m", "y"]
plt.bar(x,y, width=0.4, color="r", label="Popularity")
plt.legend()
plt.show()
```

## simple bar chart

```python
x = ["Python", "C", "C++", "Java"]
y = [85,73,60,82]
z = [20,40,30,25]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("MoonSolution",size=20)
# c = ["g", "b", "m", "y"]
plt.bar(x,y, width=0.4, color="r", label="Popularity")
plt.bar(x,z, width=0.4, color="y", label="Popularity1")
plt.legend()
plt.show()
```

## Stacked Bar Chart

```python
x = ["Python", "C", "C++", "Java"]
y = [85,73,60,82]
z = [20,40,30,25]

width=0.2
p = np.arange(len(x))
p1= [j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("MoonSolution",size=20)
# c = ["g", "b", "m", "y"]

plt.bar(p,y, width, color="g", label="Popularity")
plt.bar(p1,z, width, color="b", label="Popularity1")

plt.xticks(p+width/2, x)          #used for center the language names in bar plot
plt.legend()
plt.show()
```

## Grouped Bar Chart

```python
x = ["Python", "C", "C++", "Java"]
y = [85,73,60,82]
z = [20,40,30,25]

width=0.2
p = np.arange(len(x))
p1= [j+width for j in p]

plt.xlabel("No.", fontsize=15)
plt.ylabel("Languages", fontsize=15)
plt.title("MoonSolution",size=20)
# c = ["g", "b", "m", "y"]

plt.barh(p,y, width, color="r", label="Popularity")
plt.barh(p1,z, width, color="y", label="Popularity1")

plt.yticks(p+width/2, x)          #used for center the language names in bar plot
plt.legend()
plt.show()
```

## 2. Scatter Plot

```python
day = [2,1,3,4,5,3,6]
no = [3,5,2,6,1,3,2]
# c = ["r","g","b","m","y"]
plt.scatter(day,no, color="r")
plt.title("Scatter Plot", size=20)
plt.xlabel("Day", size=15)
plt.ylabel("No.", size=15)
plt.show()
```

### Simple Scatter Chart

```python
day = [2,1,3,4,5,3,6]
no = [3,5,2,6,1,3,2]
c = ["r","g","b","m","y","g","r"]
plt.scatter(day,no, color=c)
plt.title("Scatter Plot", size=20)
plt.xlabel("Day", size=15)
plt.ylabel("No.", size=15)
plt.show()
```

### Simple Scattle Chart with styling

```python
day = [2,1,3,4,5,3,6]
no = [3,5,2,6,1,3,2]
colors = [13,89,45,61,29,73,40]
sizes = [130,260,410,140,525,350,220]
plt.scatter(day,no, c=colors, s=sizes, marker="*", cmap="viridis")
t=plt.colorbar()
t.set_label("ColorBar", size=15)
plt.title("Scatter Plot", size=20)
plt.xlabel("Day", size=15)
plt.ylabel("No.", size=15)
plt.show()
```

```python
day = [2,1,3,4,5,7,6]
no = [3,5,2,6,1,3,2]
no1 = [2,7,3,1,6,4,5]
colors = [13,89,45,61,29,73,40]
sizes = [130,260,410,140,525,350,220]
plt.scatter(day,no, c=colors, s=sizes, cmap="viridis")
plt.scatter(day,no1, c=colors, s=sizes)
t=plt.colorbar()
t.set_label("ColorBar", size=15)
plt.title("Scatter Plot", size=20)
plt.xlabel("Day", size=15)
plt.ylabel("No.", size=15)
plt.show()
```

# Stack Plot

```python
x = [1,2,3,4,5,6]
area1= [2,3,5,4,6,5]
area2= [3,2,4,2,6,5]
area3= [1,3,2,4,5,3]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3, labels=l, colors=["r","g","b"])
plt.title("Python", size=20)
plt.xlabel("x-axis", size=15)
plt.ylabel("y-axis", size=15)
plt.legend()
plt.show()
```

# Step Plot

```python
x = [2,4,6,8,10,7]
y = [1,3,5,7,9,5]
plt.step(x,y, marker="o", color="r", ms=10, mfc="b", label="python")
plt.title("Python", size=20)
plt.xlabel("x - axis", size=15)
plt.ylabel("y - axis", size=15)
plt.legend()
plt.grid()
plt.show()
```

# Fill_Between Plot

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g", where=(x>=2) & (x<=4),alpha=0.3)
plt.title("Python",size=20)
plt.xlabel("x-axis",size=15)
plt.ylabel("y-axis",size=15)
plt.show()
```
